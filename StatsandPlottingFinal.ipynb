```python
stats.ttest_1samp(data['VIQ'], 0)  
```

The code above performs a single variable t-test on the mean VIQ scores for the entirety of the dataframe 
to determine if the mean is significantly different from 0. The p-value being equal to
1.329e-28 (<0.01) shows that the difference is significant. the variable value of 30.088 shows
that the average VIQ score is likely not 0.

The code below performs a two variable, unpaired t-test between the average VIQ scores between males and females
to determine if the two means are significantly different from eachother.
The p-value being equal to 0.445 (>0.01) shows that the diffrence is not significant.

```python
female_viq = data[data['Gender'] == 'Female']['VIQ']
male_viq = data[data['Gender'] == 'Male']['VIQ']
stats.ttest_ind(female_viq, male_viq) 
```

```python
stats.ttest_ind(data['FSIQ'], data['PIQ'])   
```

The code above performs a two variable, unpaired t-test to determine if the average FSIQ and PIQ scores for
the dataframe are significantly different. The p-value having a value of 0.643 (>0.01) shows that the
difference is not significant. This unpaired test doesn't work here, however, as unpaired tests assume the
vaules come from different groups. These variables are being measured from the same people, thus a paired t-test
is required.

The code below performs a two-variable, paired t-test to determine if the difference in FSIQ and PIQ score means are significantly
different. The p-value having a value of 0.0822 (>0.01) shows that the result is not significant.

```python
stats.ttest_rel(data['FSIQ'], data['PIQ'])  
```

```python
stats.ttest_1samp(data['FSIQ'] - data['PIQ'], 0)   
```

The code above performs a single variable t-test that measures whether the difference in FSIQ and PIQ scores is significantly
different from 0. This is equivalent to the two-variable, paired test.

```python
import numpy as np
x = np.linspace(-5, 5, 20)
print(x)
np.random.seed(1)
y = -5 + 3*x + 4 * np.random.normal(size=x.shape)
df = pandas.DataFrame({'x': x, 'y': y})
```

The code above generates a data frame with the x variable comprising 20 evenly 
spaced values between -5 and 5 made using the linspace function, and the y values 
generated from a linear equation that includes normally distributed psuedo-random noise to
simulate real data.


The code below calls the ols function from the statsmodels.formula.api package and uses 
it to generate a linear fit model from the previous dataframe.

```python
from statsmodels.formula.api import ols
model = ols("y ~ x", df).fit()
```

```python
print(model.summary())
```

The code above prints a summary of the data generated from the linear model,
as well as the summary statistics like p values, t-test values, etc.

The code below searches for and prints all attributes the linear model has in order
to be able to call those attributes when needed.

```python
for attr in dir(model):
    if not attr.startswith('_'):
        print(attr)
```

```python
print(model.params)
```

The code above prints the y intercept (-5.533) and slope (2.937) of the linear model.

The code below generate a two variable data frame from the original brain_size CSV file that consists of 
the VIQ scores for each entry and a Gender type varaible where the type is either male or female. This is done by
creating separate data frames that consist of all values of FSIQ or PIQ respectively, then concatenating the 
two data frames together.

```python
data_male = pandas.DataFrame({'Viq': data[data['Gender'] == 'Male']['VIQ'], 'typ': 'Male'})
data_female = pandas.DataFrame({'Viq': data[data['Gender'] == 'Female']['VIQ'], 'typ': 'Female'})
data_long = pandas.concat((data_male, data_female))
print(data_long) 
```

```python
model = ols("Viq ~ typ", data_long).fit()
print(model.summary()) 
```

The code above generates a linear model for the VIQ scores by Gender dataframe and prints out a summary for it.

The code below performs an f test on the linear model to determine if the variance in VIQ 
scores differs significantly between males and females. This is done by creating an array containing the
slope and intercept of the model, then using that to perform the f test on the model. The p-value of the test is
0.445 (>0.01) which means the difference in variance is not significant.

```python
A = np.array(model.params)
print(A)
print(model.f_test([0,1]))
```
